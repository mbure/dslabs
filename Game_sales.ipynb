```python
# Social media presence count by Publisher
%%sql
select Name, Platform,Publisher, count(Publisher)
from Games
group by Name
order by count(Publisher) desc
limit 20
```

```python
%%sql
select Name, Platform,Publisher, count(Publisher)
from Games
group by Name
order by count(Publisher) asc
limit 20
```

```python
#socialmedia presence by platform
%%sql
select Name, Platform,Publisher, count(Platform)
from Games
group by Name
order by count(Platform) desc
limit 20
```

```python
%%sql
select Name, Platform,Publisher, count(Platform)
from Games
group by Name
order by count(Platform) asc
limit 20
```

```python
# social media presence by Platform and Publisher.
%%sql
select count(Platform), count(Publisher), Name
from Games
where Platform >= 8 and Publisher >= 8
group by Name
order by count(Publisher) desc
limit 10;
```

```python
# Top ten performing games in North America.
%%sql
select NA_Sales, Name
from Games
group by Name
order by NA_Sales desc
limit 10;
```
