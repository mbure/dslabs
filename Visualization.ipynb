```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
print(sns.get_dataset_names())
crash_df=sns.load_dataset('car_crashes')
sns.jointplot(x='speeding',y='alcohol', data=crash_df, kind='scatter')
```

```python
sns.kdeplot(crash_df['alcohol'])
# visualize the relationship between tipping,total bill and smokers
tips_df=sns.load_dataset('tips')
tips_df
sns.barplot(x='sex',y='total_bill',data=tips_df)
```

```python
sns.countplot(x='sex', data=tips_df)
```

```python
sns.boxplot(x='day',y='total_bill', hue='sex', data=tips_df)
```

```python
sns.stripplot(x='day',y='total_bill',hue='sex',data=tips_df)
```

```python
plt.legend(loc=0)
plt.figure(figsize=(8,6))
```

```python
crash_mx=crash_df.corr()
sns.heatmap(crash_mx,annot=True,cmap='Greens')

 
```
